# Introduction
the Fourier Transform as signular standing in signal processing because the the way sampling imposes a bandwith-centric view.   
The Discrete Fourier Transform (DFT) is the primary analysis tool for this perspective.

This is the DFT matrix
$$
\mathbb{U} = \frac{1}{\sqrt{N}} \left[ \exp \left( j \frac{2\pi}{N} nk \right) \right]_{n \in \{ 0, N_s - 1\}, k \in \{ 0, N-1\}}
$$
where $n$ countsd through the number of samples and $k$ indexes the discrete frequencies as columns in the matrix.

```python
import numpy as np
import math
import matplotlib.pyplot as plt

N = 100
L = 20
s = np.zeros(N - 1)
for k in np.arange(1, 300, 2):
    s += 1/float(k) * np.sin( 2* math.pi * k / L * np.arange(1,N,1))
plt.plot(s)
plt.title('Sum of first {} terms'.format( (k-1)/2 + 1))
plt.show()
```

```python
def sfou_exp(Km):
    """ what is this ? """
    K_max = int(Km)
    L = 400
    N = 1000
    k = 0
    s = np.zeros(N - 1)
    for k in np.arange(1, K_max):
        ak = 0
        bk = 1.0 / k if (k % 2) == 1 else 0
        s += ak * np.cos(2* math.pi * k / L * np.arange(1,N,1)) + bk * np.sin(2* math.pi * k / L * np.arange(1,N,1))
    ax = plt.axes(xlim=(0, N), ylim=(-2,2))
    ax.plot(s)
    plt.title('Sum with {} terms'.format(k+1))
    plt.show()
    
sfou_exp(5)
```
