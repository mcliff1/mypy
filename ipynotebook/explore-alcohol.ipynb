```python
# data from https://github.com/fivethirtyeight/data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('../../data/alcohol-consumption/drinks.csv')
df.head()
```

```python
df.describe()
```

```python
# show the top by beer_servings
beer = df.sort_values(by='beer_servings', ascending=False)['country']
wine = df.sort_values(by='wine_servings', ascending=False)['country']
spirit = df.sort_values(by='spirit_servings', ascending=False)['country']
alcohol = df.sort_values(by='total_litres_of_pure_alcohol', ascending=False)['country']
```

```python
beer.head()
```

```python
df3 = pd.DataFrame(data={'beer': beer.reset_index(drop=True), 
                         'sprit': spirit.reset_index(drop=True), 
                         'wine': wine.reset_index(drop=True),
                         'alcohol': alcohol.reset_index(drop=True)})
df3.head(10)
```
