```python
# plot the sigmoid
def sigma(x):
    return 1/ (1 + np.exp(-x))

x = np.linspace(-5, 5, 100)
plt.plot(x, sigma(x), 'b')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid()
plt.text(-4, 0.85, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=16)
plt.show()
```
