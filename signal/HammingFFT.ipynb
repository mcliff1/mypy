```python
import numpy as np
import matplotlib.pyplot as plt

window = np.hamming(20)
plt.plot(window)
plt.show()
```

```python
A = np.fft.fft(window, 2048) / 25.5
mag = np.abs(np.fft.fftshift(A))
freq = np.linspace(-0.5, 0.5, len(A))
resp = np.clip(20* np.log10(mag), -100, 100)
plt.plot(freq, resp)
plt.axis('tight')
plt.show()
```
